```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("ckd.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#eliminate an unwanted columns
#dataset=dataset.drop("column name", axis=1) Default axis =0 (row) 1 (column)
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes', ]]
depen=dataset[['classification_yes']]
```

```python
indep.shape
```

```python
dataset.dtypes
```

```python
depen.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,depen,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid={'penalty':['l2'],'solver':['lbfgs', 'liblinear', 'newton-cg', 'saga']}
grid=GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
grid.fit(x_train, y_train)
```

```python
#print best parameter after tunning
print(grid.best_params_)
```

```python
#print the result
re=grid.cv_results_
print(re)
```

```python
grid_predictions=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, grid_predictions)
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
```

```python
print("The Confusion Matrix:\n", cm)
```

```python
print("The report:\n", clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
```
