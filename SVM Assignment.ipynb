```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("ckd.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes', ]]
depen=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,depen,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid={'gamma':['auto','scale'],'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[10,100,1000,2000,3000]}
grid=GridSearchCV(SVC(probability=True), param_grid, refit=True, verbose=3, cv=5, n_jobs=-1, scoring='f1_weighted')
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
```

```python
re=grid.cv_results_
print (re)
```

```python
grid_predictions=grid.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, grid_predictions)
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
```

```python
print("The Confusion Matrix:\n", cm)
```

```python
print("The Report:/n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
```
